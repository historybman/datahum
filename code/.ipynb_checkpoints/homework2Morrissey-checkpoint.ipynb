Make a scatter plot of the relationship between raw budget and domestic receipts.  Just to visualize the possible coorelation.  Then do the pearson test.

````python
bechdel_data.plot(kind = 'scatter', x = 'budget', y = 'domgross', color = 'r', marker = 'o')
plt.show()
````

Now find the pearson coefficient to see what the correlation is.  

````python
from scipy.stats.stats import pearsonr
coef = pearsonr(bechdel_data['domgross'], bechdel_data['budget'])
print(coef)
````
