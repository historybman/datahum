```python
import pandas as pd
pd.set_option('display.mpl_style', 'default')
```

```python
montreal = pd.read_csv('../data/Montreal/comptagevelo2012.csv', parse_dates=['Date'], dayfirst=True, index_col='Date')
```

```python
montreal['Berri1'].plot()
```

```python
berri_bikes = montreal[['Berri1']]
```

```python
berri_bikes[:5]
```

```python
berri_bikes.index.weekday
```

```python
berri_bikes['weekday'] = berri_bikes.index.weekday
berri_bikes[:5]
```

```python
weekday_counts = berri_bikes.groupby('weekday').aggregate(sum)
weekday_counts
```

```python
weekday_counts.index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekday_counts
```

```python
weekday_counts.plot(kind = 'bar')
```
